## Pre processing

Liste des fonctions

```python
def joindre_deux_Dataframe(df1,df2):
    """joindre deux Dataframe

    Args:
        df1 (Dataframe): df1
        df2 (Dataframe): df2

    Returns:
        Dataframe: _description_
    """
    jointure = pd.merge(df1,df2,how='inner')
    return jointure
```

```python
def suprimmer_les_doublons(liste_datas):
    """suprimmer_les_doublons

    Args:
        liste_datas (liste): listes des Dataframes
    """
    for data in liste_datas:
        data.drop_duplicates()
```

```python
def calculer_les_indices_de_dispertions(data,target):
    """Calculer les indices de dispertions
    Args:
        data (Dataframe): dataset √† traiter
        target (str): colonne √† traiter

    Returns:
        tuple: indice_dispertion_positive,indice_dispertion_n√©gative
    """
    indice_dispertion_positive = 1.5 * (data[target].quantile(q=0.75 - 0.25))
    indice_dispertion_n√©gative = -1.5 * (data[target].quantile(q=0.75 - 0.25))
    return indice_dispertion_positive,indice_dispertion_n√©gative
```

```python
def cr√©er_un_sous_ensemble_de_donn√©es(data,target):
    """Calculer les indices de dispertions
    Args:
        data (Dataframe): dataset √† traiter
        target (str): colonne √† traiter

    Returns:
        Dataframe: indice_dispertion_positive,indice_dispertion_n√©gative
    """
    indice_dispertion_positive,indice_dispertion_n√©gative = calculer_les_indices_de_dispertions(data,target)
    data_sub = data[~((data[target] > indice_dispertion_positive) | (data[target] < indice_dispertion_n√©gative))] 
    return data_sub
```

```python
def remplacer_les_valeurs_extr√®mes_par_des_NaN(data,target):
    """Remplacer les valeurs extr√®mes par des NaN
    Args:
        data (Dataframe): dataset √† traiter
        target (str): colonne √† traiter
    """
    indice_dispertion_positive,indice_dispertion_n√©gative = calculer_les_indices_de_dispertions(data,target)
    condition = (data[target] > indice_dispertion_positive) | (data[target] < indice_dispertion_n√©gative)
    data.loc[condition, target] = np.nan
```

```python
def d√©coupage_train_test(data,target,feature):
    """d√©coupage train test
    Args:
        data (Dataframe): dataset √† traiter
        target (str): colonne √† traiter
    Returns:
        tuple: X_train, X_test, y_train, y_test
    """
    x = data[target].to_numpy()
    y = data[feature].to_numpy()
    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
    return X_train, X_test, y_train, y_test
```

```python
def entrainer_r√©gression_lin√©aire_et_pr√©dire(data,target,feature):
    """Entrainer r√©gression lin√©aire et pr√©dire
    Args:
        data (Dataframe): dataset √† traiter
        target (str): colonne √† traiter
    Returns:
        tuple: coeffs,y_pred, mse
    """
    X, y, = d√©coupage_train_test(data,target,feature)[0], d√©coupage_train_test(data,target,feature)[2]
    coeffs = np.polyfit(X, y, 1)
    y_pred = np.polyval(coeffs,X)
    y_pred = pd.Series(y_pred)
    mse = np.mean(y - y_pred)**2
    fig = px.scatter(
    x=X, y=y_pred, opacity=0.65,
    trendline='ols', trendline_color_override='darkblue'
    )
    fig.show()
    return coeffs,y_pred, mse
```

```python
def tester_r√©gression_lin√©aire_et_pr√©dire(data,target,feature,coeffs):
    """Tester r√©gression lin√©aire et pr√©dire
      Args:
        data (Dataframe): dataset √† traiter
        target (str): colonne √† traiter
    Returns:
        tuple: y_pred, mse
    """
    X, y, = d√©coupage_train_test(data,target,feature)[1], d√©coupage_train_test(data,target,feature)[3]
    y_pred = np.polyval(coeffs,X)
    y_pred = pd.Series(y_pred)
    mse = np.mean(y - y_pred)**2
    fig = px.scatter(
    x=X, y=y_pred, opacity=0.65,
    trendline='ols', trendline_color_override='darkblue'
    )
    fig.show()
    return y_pred, mse
```

```python
def remplacer_les_NaN_par_des_pr√©dictions(data,target,feature,coeffs):
    """remplacer les NaN par des pr√©dictions
      Args:
        data (Dataframe): dataset √† traiter
        target (str): colonne √† traiter
        feature (str): colonne √† traiter
        coeffs (ndarray): coefficients a et b
    """
    predictions = np.polyval(coeffs, data[feature])
    predictions = pd.Series(data=predictions)
    data.fillna({target: predictions}, inplace=True)
```

```python
def afficher_graphique_corr√©lation(data):
    """afficher graphique corr√©lation
      Args:
        data (Dataframe): dataset √† traiter
    """
    fig = px.imshow(data.corr(), text_auto=True)
    fig.show()
```

```python
def normalize_data(data, min_range=0, max_range=1):
    """
    Normalizes a given pandas Series or DataFrame column.
    Normalise les donn√©es d'une colonne du DataFrame
    Args :
    data (pandas.Series or pandas.DataFrame): The data to be normalized.
    min_range (float): Minimum de valeur √† d√©sirer. Default is 0.
    max_range (float): Maximum de valeur √† d√©sirer Default is 1.

    Returns:
    Normalized data.
    """
    min_val = data.min()
    max_val = data.max()
    normalized_data = (data - min_val) * (max_range - min_range) / (max_val - min_val) + min_range
    return normalized_data
```
